'' Q1. What are the most common reasons for missing data in ETL pipelines? ''

Ans1. The most common reasons for missing data in ETL pipelines are 
                              
        Schema Drift: Upstream teams change columns without notice.
    
        Network Failures: API timeouts or connection drops halt extraction.
        
        Bad Logic: Faulty inner joins or strict filters drop rows.
        
        Delayed Files: Sources deliver data after the pipeline runs.
        
        Poor Error Handling: Bad records crash jobs or get dropped.

Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?

Ans2. It is considered as bad practice because : 

       Skews Metrics: Deleting rows distorts averages, sums, and KPIs.
           
        Hides Bugs: It masks underlying pipeline or system failures.
            
        Introduces Bias: Missing data is rarely random; deletion skews demographics.
            
        Breaks Tracking: Removing partial records breaks historical audit trails.
            
        Loses Context: Optional fields (like "Notes") often contain valuable blanks.

Q4. Why is median imputation preferred over mean imputation for skewed data such as income?

Ans4. Median imputation is preferred for skewed data because it resists outliers, whereas the mean gets pulled toward extreme values.

Q5. What is forward fill and in what type of dataset is it most useful?

Ans5. Forward fill (ffill) propagates the last known non-null value forward to replace subsequent missing values until a new valid value appears.

Best Dataset Types

Time-Series Data: Sensor readings, stock prices, or smart meter logs.

Sequential Data: High-frequency logs where variables stay constant between updates.

Scenario: A temperature sensor records 22°C at 1:00 PM but drops connection at 1:05 PM; forward fill assumes it stays 22°C until the next reading

Q6. Why should flagging missing values be done before imputation in an ETL workflow?

Ans6. Flagging missing values before imputation preserves the original data state and protects downstream model performance.

Q7. Consider a scenario where income is missing for many customers. 

 How can this missingness itself provide business insights?

And6. Missingness in income data often signals customer friction, behavioral traits, or demographic segments.

Business Insights

Process Friction: High dropouts on the income field reveal a intrusive sign-up form.

Privacy Concerns: Reluctance to share income points to low brand trust.

Customer Segment: Low-income or gig-economy workers often hide income due to qualification fears.

Device/Channel Issues: Drops on mobile screens suggest the input field is broken or poorly formatted.

Employment Shifts: Blanks can flag retirees, students, or homemakers who lack a traditional salary.
